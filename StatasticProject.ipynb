2.B. Probability that exactly one of the items is defective? 
2.C. Probability that two or fewer of the items are defective? 
2.D. Probability that three or more of the items are defective ? 

```python
#Defect 5% probability
#sample size =10
from scipy.stats import binom
p=5/100
n=10
k=np.arange(0,10)
binomial=binom.pmf(k,n,p)
binomial
```

```python
print('none of the item is defected=', binomial[0])
```

```python
print('Exactly 1 item is defective',binomial[1])
```

```python
#probabilty defect<=2
defect2= binomial[0] + binomial[1]+ binomial[2]
print('two or fewer of the items are defective', defect2)
```

```python
defect3= 1- (binomial[0] + binomial[1]+ binomial[2])
print('three or more of the items are defective', defect3)
```
